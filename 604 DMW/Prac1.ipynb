```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
data = pd.read_csv("data.csv")
data
```

```python
x=data['tcs']
y=data['wipro']
```

```python
n=stats.linregress(x,y)
print(n)
```

```python
s=n.slope
print("slope",s)
i =n.intercept
print("Intercept",i)
```

```python
pred_y = s*x + i
print(pred_y)
```

```python
r_squared = n.rvalue
print(r_squared)
```
